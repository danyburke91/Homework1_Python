<p>Cube Number Test... Print out all cubed numbers up to the total value 1000. Meaning that if the cubed number is over 1000 break the loop.</p>

```python
cubed_list = []
for num in range(1,11):
    cubed_num = num**3
    cubed_list.append(cubed_num)
print(cubed_num)

```
